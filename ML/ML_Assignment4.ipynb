Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
x_list=[]
fx_list=[]
def gradient(start,learning_rate,iterations):
    x=start
    x_list.append(x)
    fx_list.append(objective(x))
    print(f"Iteration 0 -> x = {x}    f(x) = {objective(x)}")
    for i in range(iterations):
        der=derivative(x)
        x=x-(learning_rate*der)
        x_list.append(x)
        fx_list.append(objective(x))
        print(f"Iteration {i+1} -> x = {x}          f(x) = {objective(x)}")
```

```python
start=2
learning_rate=0.1
iterations=30
gradient(start,learning_rate,iterations)
```

```python
x_val=np.linspace(-15,15,100)
plt.plot(x_val,objective(x_val))
plt.plot(2,objective(2),"ro")
```

```python
x_val=np.linspace(-5,5,100)
plt.plot(x_val,objective(x_val))
plt.plot(x_list,fx_list,'.-',color="red")
plt.plot(x_list[29],fx_list[29],'go')
print(f"Local Minima Occurs at x = {x_list[29]} and f(x) = {fx_list[29]} .")
```

```python
def objective2(x):
    return (x**3)-(3*x**2)+1
```

```python
def derivative2(x):
    return (3*x**2)-(6*x)
```

```python
x_list2=[]
fx_list2=[]
def gradient(start,learning_rate,iterations):
    x=start
    x_list2.append(x)
    fx_list2.append(objective2(x))
    print(f"Iteration 0 -> x = {x}    f(x) = {objective2(x)}")
    for i in range(iterations):
        der=derivative2(x)
        x=x-(learning_rate*der)
        x_list2.append(x)
        fx_list2.append(objective2(x))
        print(f"Iteration {i+1} -> x = {x}          f(x) = {objective2(x)}")
```

```python
start=1
learning_rate=0.1
iterations=50
gradient(start,learning_rate,iterations)
```

```python
x_val=np.linspace(-15,15,100)
plt.plot(x_val,objective2(x_val))
plt.plot(1,objective2(1),"ro")
```
